```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("cost_data.csv")

# --- Daily cost plot ---
plt.figure(figsize=(8, 5))
plt.plot(df["queries_per_day"], df["ondemand_cost_day_usd"],
         marker="o", label="On-demand (day)")
plt.plot(df["queries_per_day"], df["provisioned_cost_day_usd"],
         marker="s", label="Provisioned (day)")

plt.xlabel("Queries per day")
plt.ylabel("Cost per day (USD)")
plt.title("Bedrock Mistral: On-demand vs Provisioned (Daily)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# --- Monthly cost plot ---
plt.figure(figsize=(8, 5))
plt.plot(df["queries_per_day"], df["ondemand_cost_month_usd"],
         marker="o", label="On-demand (month)")
plt.plot(df["queries_per_day"], df["provisioned_cost_month_usd"],
         marker="s", label="Provisioned (month)")

plt.xlabel("Queries per day")
plt.ylabel("Cost per month (USD)")
plt.title("Bedrock Mistral: On-demand vs Provisioned (Monthly)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio

# Force browser renderer (avoids nbformat / notebook MIME)
pio.renderers.default = "browser"

df = pd.read_csv("cost_data.csv")

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df["queries_per_day"],
    y=df["ondemand_cost_day_usd"],
    mode="lines+markers",
    name="On-demand (day)",
    hovertemplate="Queries: %{x}<br>Cost/day: $%{y:.2f}<extra></extra>",
))

fig.add_trace(go.Scatter(
    x=df["queries_per_day"],
    y=df["provisioned_cost_day_usd"],
    mode="lines+markers",
    name="Provisioned (day)",
    hovertemplate="Queries: %{x}<br>Cost/day: $%{y:.2f}<extra></extra>",
))

fig.add_trace(go.Scatter(
    x=df["queries_per_day"],
    y=df["ondemand_cost_month_usd"],
    mode="lines+markers",
    name="On-demand (month)",
    hovertemplate="Queries: %{x}<br>Cost/month: $%{y:.2f}<extra></extra>",
))

fig.add_trace(go.Scatter(
    x=df["queries_per_day"],
    y=df["provisioned_cost_month_usd"],
    mode="lines+markers",
    name="Provisioned (month)",
    hovertemplate="Queries: %{x}<br>Cost/month: $%{y:.2f}<extra></extra>",
))

fig.update_layout(
    title="Bedrock Mistral: On-demand vs Provisioned (Daily & Monthly)",
    xaxis_title="Queries per day",
    yaxis_title="Cost (USD)",
    hovermode="x unified",
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio

# Force browser renderer so it doesn't depend on notebook MIME
pio.renderers.default = "browser"

# Load your CSV
df = pd.read_csv("cost_data.csv")

fig = go.Figure()

# On-demand monthly cost
fig.add_trace(go.Scatter(
    x=df["queries_per_day"],
    y=df["ondemand_cost_month_usd"],
    mode="lines+markers",
    name="On-demand (month)",
    hovertemplate="Queries/day: %{x}<br>On-demand/month: $%{y:.2f}<extra></extra>",
))

# Provisioned monthly cost
fig.add_trace(go.Scatter(
    x=df["queries_per_day"],
    y=df["provisioned_cost_month_usd"],
    mode="lines+markers",
    name="Provisioned throughput (month)",
    hovertemplate="Queries/day: %{x}<br>Throughput/month: $%{y:.2f}<extra></extra>",
))

fig.update_layout(
    title="Monthly Cost: On-demand vs Provisioned Throughput",
    xaxis_title="Queries per day",
    yaxis_title="Cost per month (USD)",
    hovermode="x unified",
    legend=dict(x=0.01, y=0.99),
)

fig.show()
```
